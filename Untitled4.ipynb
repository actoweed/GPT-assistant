```python
from sentence_transformers import SentenceTransformer
from pinecone import Pinecone, ServerlessSpec
import numpy as np
import os
from openai import OpenAI
import torch

# Проверяем доступность GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Используемое устройство: {device}")

os.environ["PINECONE_API_KEY"] = "INSERT YOUR PINECONE API KEY"

client = OpenAI(api_key="YOUR OPENAI API KEY")

pc = Pinecone(api_key=os.environ["PINECONE_API_KEY"])

index_name = "multilingual-e5-large"
if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=384,
        metric="cosine",
        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        )
    )

index = pc.Index(index_name)

embedding_model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

def generate_response(query, context):
    prompt = f"На основе следующего контекста ответь на вопрос: {query}. Контекст: {context}"
    response = client.chat.completions.create(
        model="gpt-4o-2024-11-20",
        messages=[
            {"role": "system", "content": "Ты — помощник, который отвечает на вопросы на основе предоставленного контекста."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=500,
        temperature=0.6,
    )
    return response.choices[0].message.content

def load_txt_files(file_paths):
    documents = []
    for file_path in file_paths:
        with open(file_path, 'r', encoding='utf-8') as file:
            documents.append(file.read())
    return documents

file_paths = ["/etc/TXT FILES/111.txt"]
documents = load_txt_files(file_paths)

documents = [doc.encode('utf-8').decode('utf-8') for doc in documents]

embeddings = embedding_model.encode(documents)
print(f"Размерность эмбеддингов: {embeddings.shape}")

for i, (doc, embedding) in enumerate(zip(documents, embeddings)):
    index.upsert(
        vectors=[
            {
                "id": str(i),
                "values": embedding.tolist(),
                "metadata": {"text": doc}
            }
        ]
    )

print("Данные успешно загружены в Pinecone!")

query = input("Введите ваш запрос: ")
query_embedding = embedding_model.encode(query).tolist()

results = index.query(
    vector=query_embedding,
    top_k=3,
    include_metadata=True
)

found_texts = [match["metadata"]["text"] for match in results["matches"]]

context = " ".join(found_texts)

# Генерация ответа
response = generate_response(query, context)

# Вывод ответа
print("\nОтвет:")
print(response)
```
